# 1. Obtain and Filter the Data

```python
linear_boston = boston_housing.filter(["Per_Capita_Crimerate", "Charlsriver", "Average_Rooms_dwelling", "MEDValue_owner"])
linear_boston.head()
```

```python
linear_boston.shape
```

```python
fig, ax = plt.subplots()
ax.hist(linear_boston["MEDValue_owner"]);
```

```python

fig, ((ax1,ax2,ax3)) = plt.subplots(nrows = 1,
                                          ncols = 3,
                                          figsize = (30,5))
ax1.hist(linear_boston["Per_Capita_Crimerate"])
ax1.set(title ="Crime percentage",
      xlabel = "Per_Capita_Crimerate",
      ylabel = "Percentage")
ax2.hist(linear_boston["Average_Rooms_dwelling"])
ax2.set(title ="Room_dwelling",
        xlabel = "Average rooms",
        ylabel = "Range")
ax3.plot(linear_boston["MEDValue_owner"]);
ax3.set(title ="Price",
      xlabel = "Median Value",
      ylabel = "Value");
```

# 2. Preparing the model - Linear Model

```python
from sklearn.model_selection import train_test_split

# Dependent variable

y = linear_boston["MEDValue_owner"]

# Independent variable

x = linear_boston.iloc[:,0:3]

#split the data into a training set and a test set

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state =0)

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
method1 = LinearRegression()
method1.fit(x_train, y_train)

print('labels:\n', x.columns)
print('Coefficients:\n', method1.coef_)
print('Intercept: \n', method1.intercept_)
print('R2 for Train', method1.score(x_train, y_train))
print('R2 for Test(cross validation)', method1.score(x_test, y_test))
```

```python
pd.DataFrame(list(zip(x.columns, method1.coef_)),
            columns = ['features', 'estimateCoeffs'])
```

```python
y_pred = method1.predict(x_test)
print(y_pred)
y_pred.shape
```

```python
print(y_test)
```

```python
# checking the accuracy using mean squared error

from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, y_pred))
```
